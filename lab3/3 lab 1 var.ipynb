```python
def psi(i, x):
    if(i == N):
        return x/h - (i - 1)
    if(x >= (i-1)*h and x <= i*h):
        return x/h - (i - 1)
    elif(x >= i*h and x <= (i + 1)*h):
        return -x/h + (i + 1)
    else:
        return 0
```

```python
def psi_x(i, x):
    if(i == N):
        return 1/h
    if(x >= (i-1)*h and x <= i*h):
        return 1/h
    elif(x >= i*h and x <= (i + 1)*h):
        return -1/h
    else:
        return 0
```
